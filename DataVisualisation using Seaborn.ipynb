```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Apart from darkgrid there are other styles that you can use like:-

#### darkgrid
#### whitegrid
#### dark
#### white
#### ticks 

```python
sns.set_style('darkgrid')
df = pd.read_csv('data.csv')
```

# sns.scatterplot()

```python
df.head()
```

```python
plt.figure(figsize = (10,6))  # for figure size, Hue for differencing between variables
sns.scatterplot(x = df['Primary_Only'], y = df['HrSec_Only'], hue = df["year"], alpha = 0.8) # alpha for visibility of plots 
plt.show()
```

# sns.countplot()

```python
plt.figure(figsize = (9,6))
sns.countplot(x = "year", data = df)
plt.show()
```

# sns.distplot()

```python
plt.figure(figsize = (9,6))
sns.distplot(df['Primary_Only'], kde = False)
plt.show()
```

# sns.kdeplot()

```python
plt.figure(figsize = (9,6))
sns.kdeplot(df['Primary_Only'])
plt.show()
```

# sns.regplot()

```python
plt.figure(figsize = (9,6))
sns.regplot(x = df['Primary_Only'], 
            y = df['Sec_with_HrSec.'],
            scatter_kws = {'color':'pink'},
            line_kws = {'color':'black'}
)
```

# sns.lmplot()

```python
plt.figure(figsize = (9,6))
sns.lmplot(x = 'Primary_Only', 
           y = 'Sec_with_HrSec.',
           hue = 'year',
           data = df
)
```

# sns.pairplot()

```python
sns.pairplot(df[['Primary_Only', 'Sec_with_HrSec.', 'U_Primary_With_Sec']])
```

# sns.jointplot()

```python
sns.jointplot(x = 'Primary_Only', 
              y = 'Sec_with_HrSec.',
              data = df
             
)
plt.show()
```

# sns.boxplot()

```python
sns.boxplot(x = 'Primary_Only', 
            data = df
)
plt.show()
```

```python
plt.figure(figsize = (25,6))
sns.boxplot(data = df)
plt.show()
```

# sns.swarmplot()

```python
plt.figure(figsize = (20,8))
sns.swarmplot(x = 'Primary_Only',
              y = 'U_Primary_Only', 
              data = df,
              alpha = 0.8
)
plt.show()
```

# sns.heatmap()

```python
sns.heatmap(df.corr(),
            annot = True,
            cmap = 'inferno')
plt.show()
```

# Color Pallete

```python
plt.figure(figsize = (20,8))
colors = [
           '#F8D030',
           '#E0C068',
           '#EE99AC',
           '#C03028',
           '#F85888'
]
sns.stripplot(x = 'State_UT',
              data = df,
              palette = colors
)
plt.show()
```
